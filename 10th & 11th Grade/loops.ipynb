## ฺฉูุชุฑู ุฌุฑุงู ุฏุฑ ุญููู: `break`, `continue`, `pass`
- `break`: ุฎุฑูุฌ ููุฑ ุงุฒ ุญููู  
- `continue`: ูพุฑุด ุจู ุชฺฉุฑุงุฑ ุจุนุฏ  
- `pass`: ุงูุฌุงู ูุฏุงุฏู ูฺโฺฉุงุฑ (ุฌุงโูฺฏูุฏุงุฑ ุขููุฒุด)


```python

# break, continue, pass demo
for i in range(1, 8):
    if i == 2:
        pass  # placeholder; does nothing
    if i == 4:
        continue  # skip printing 4
    if i == 6:
        break     # stop the loop entirely
    print("i =", i)
```


### ุณุงุฎุชุงุฑ ฺฉูุชุฑุดูุงุฎุชูโุดุฏู: `for/else` ู `while/else`

```python

# for-else example: search for a target
nums = [3, 7, 10, 12]
target = 8

for n in nums:
    if n == target:
        print("Found!")
        break
else:
    print("Not found (loop completed without break).")
```

## ุงุนุชุจุงุฑุณูุฌ ูุฑูุฏ ุจุง `while True`
ุฑูุด ุงุณุชุงูุฏุงุฑุฏ ุจุฑุง ุชฺฉุฑุงุฑู ุฏุฑุฎูุงุณุช ุชุง ููุช ูุฑูุฏ ูุนุชุจุฑ ุฏุฑุงูุช ุดูุฏ.


```python

# Input validation loop
while True:
    user = input("Enter a positive integer (or 'q' to quit): ")
    if user.lower() == 'q':
        print("Quit requested.")
        break
    try:
        val = int(user)
        if val > 0:
            print("Thanks! You entered:", val)
            break
        else:
            print("Please enter a number greater than 0.")
    except ValueError:
        print("Invalid input. Try again.")
```

## ูุซุงู ูุงูุน ฑ: ูพุฑุฏุงุฒุด ููุฑุงุช (ฐ ุชุง ฒฐ) ุจุง ุญูููโูุง
- ุดูุงุฑุด ููุฑุงุช ูุจูู ู ูุฑุฏูุฏ  
- ูุญุงุณุจูู ูุงูฺฏู  
- ุงูุชู ุจุดูู ู ฺฉููู ุจุฏูู ุชูุงุจุน ุขูุงุฏู


```python

# Scores processing (0..20)
scores = [19, 14.5, 9, 16, 11, 7, 20, 12]

# Count pass/fail
pass_count = 0
fail_count = 0
for s in scores:
    if s >= 10:
        pass_count += 1
    else:
        fail_count += 1

# Compute average
total = 0
for s in scores:
    total += s
avg = total / len(scores)

# Find max/min manually
mx = scores[0]
mn = scores[0]
for s in scores:
    if s > mx:
        mx = s
    if s < mn:
        mn = s

print(f"Scores: {scores}")
print(f"Passing: {pass_count}, Failing: {fail_count}")
print(f"Average: {avg:.2f}")
print(f"Max: {mx}, Min: {mn}")
```


## ูุซุงู ูุงูุน ฒ: ุชุดุฎุต ุงูู ุจูุฏู ุนุฏุฏ ุจุง `for/else`
ุงฺฏุฑ ุชุง ุงูุชูุง ุญููู ูฺ ููุณูู ูพุฏุง ูุดุฏุ ุนุฏุฏ **ุงูู** ุงุณุช.


```python

# Primality test using for-else
n = int(input("Enter an integer >= 2: "))

is_prime = True
for d in range(2, int(n**0.5) + 1):
    if n % d == 0:
        is_prime = False
        break
else:
    # else-part for the loop (no break)
    pass

if is_prime:
    print(f"{n} is prime")
else:
    print(f"{n} is not prime")
```


## ูุซุงู ูุงูุน ณ: ูุงฺฉุชูุฑู ู ุฏูุจุงููู ูุจููุงฺ


```python

# Factorial using while
m = int(input("Enter a non-negative integer for factorial: "))
fact = 1
i = 1
while i <= m:
    fact *= i
    i += 1
print(f"{m}! = {fact}")
```

```python

# Fibonacci using for
k = int(input("How many Fibonacci numbers to generate? "))

fib = [0, 1]
for _ in range(2, k):
    fib.append(fib[-1] + fib[-2])

print("Fibonacci:", fib[:k])
```

## (ุงุฎุชุงุฑ) ุขุดูุง ุจุง List Comprehension
ฺฏุงู ูโุชูุงูู ุจูโุฌุง ุญูููู ฺูุฏุฎุทุ ุฏุฑ ฺฉ ุฎุท ูุณุช ุจุณุงุฒูุ ุงูุง ุฏุฑ ุขููุฒุด ุงูููุ **ุฎูุงูุง** ุฑุง ูุฏุง ุงุฎุชุตุงุฑ ูฺฉูุฏ.


```python

# List comprehension vs. regular loop
# Square numbers 1..10
squares_loop = []
for i in range(1, 11):
    squares_loop.append(i*i)

squares_comp = [i*i for i in range(1, 11)]

print("Loop:", squares_loop)
print("Comprehension:", squares_comp)
```
