## ğŸ–¨ï¸ print and f-string

```python

# Basic print
print("Hello, world!")

# Printing multiple values
name = "Ali"
age = 18
print("My Name is", name, "and I am", age, "years old.")

# Using f-string
name = "Sara"
age = 20
print(f"My name is {name} and I am {age} years old.")

# Expressions inside f-strings
a = 5
b = 7
print(f"{a} + {b} = {a + b}")
```

## ğŸ”¢ str(), int(), float() â€” Casting

```python

# Type casting examples
x = "5"
print(type(x))  # str

x = int(x)
print(type(x))  # int

y = 3.14
print(str(y))   # Convert to string
print(float("7.25"))  # Convert string to float
```
